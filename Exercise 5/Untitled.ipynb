```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def bazaraa_shetty(x,y):
    return (x-2)**4 + (x-2*y)**2
def roosenbrock(x,y):
    return 100*(y-(x)**2)**2 + (1-x)**2
x = np.linspace (-2.    # lower limit
                , 3     # upper limit
                , 21  
                )
y = np.linspace (-2.    # lower limit
                , 3     # upper limit
                , 21    # generate 51 points between 0 and 3
                )
def gradient_bazaraa_shetty(x,y):
    derivative_x = 4*(x-2)**3 + 2*(x-(2*y))
    derivative_y = 2*(x-(2*y))*(-2)
    gradient = np.array([derivative_x,derivative_y], dtype=float)
    return gradient
        
def gradient_roosenbrock(x,y):
    derivative_x = 200*(-2*x)*(y-(x**2))+(2*x)-2
    derivative_y = 200*(y-(x**2))
    gradient =  np.array([derivative_x,derivative_y], dtype=float)
    return gradient
```

```python
def check_condition1(old_input,new_input,step_size,delta):
    new_function_value = float(roosenbrock(new_input[0],new_input[1]))
    old_function_value = float(roosenbrock(old_input[0],old_input[1]))
    gradient = gradient_roosenbrock(old_input[0],old_input[1])
    decent_direction = -1*(np.linalg.norm(gradient))
    check = new_function_value <= old_function_value + (step_size*delta*decent_direction)
    return check
    
def armijoLineSearch(step_size,delta,func,initial_input,direction):
    old_input = initial_input
    new_input = np.add(old_input, np.array([direction*step_size,direction*step_size],dtype=float))
    is_condition1_true = check_condition1(old_input,new_input,step_size,delta)
    current_step_size = step_size
        
    while(is_condition1_true):
        current_step_size = current_step_size*2
        old_input = new_input
        new_input = np.add(old_input, np.array([direction*current_step_size,direction*current_step_size],dtype=float))
        is_condition1_true = check_condition1(old_input,new_input,current_step_size,delta)
        #print(is_condition1_true)
        
    while not(is_condition1_true):
        current_step_size = current_step_size/2
        old_input = new_input
        new_input = np.add(old_input, np.array([direction*current_step_size,direction*current_step_size],dtype=float))
        is_condition1_true = check_condition1(old_input,new_input,current_step_size,delta)
#       print(is_condition1_true)
          
    #print('step after condition2',current_step_size)
    return current_step_size
    
```
