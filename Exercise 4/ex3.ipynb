```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import fft
from matplotlib.image import imread
from matplotlib.pyplot import imshow
```

```python
A = imread('easter.PNG')
B = np.mean(A,-1)
print(A.shape)
imshow(B)
```

```python
U,S,Vt = np.linalg.svd(B, full_matrices=True)
S = np.diag(S)
j=0

print(S[0:4,:4])

for r in (int((len(S))*.2),int((len(S))*.1),int((len(S))*.02)):
    approximation = U[:,:r] @ S[0:r,:r]@ Vt[:r,:]
    plt.figure(j)
    j +=1
    imshow(approximation)
```
