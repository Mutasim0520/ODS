```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import fft
from matplotlib.image import imread
from matplotlib.pyplot import imshow

A = imread('easter.PNG')
gs_image = np.mean(A, -1)

imshow(gs_image)
plt.axis('off')
```

```python
coef_array = fft.fft2(gs_image)[::-1]
sort_by_magnitude = np.sort(np.abs(coef_array.reshape(-1)))

for keep in (0.1,0.05,0.02,0.01):
    thresh = sort_by_magnitude[int(np.floor((1-keep)*len(sort_by_magnitude)))]
    ind = np.abs(coef_array)>thresh
    atlow = coef_array*ind
    co = fft.ifft2(atlow).real
   
    plt.figure()
    plt.imshow(co)
    plt.title(keep)

    
```
